Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Import Dataset

```python
data = pd.read_excel("trupti data science.xlsx",error_bas_lines=False)
```

Mean And Median Of The Distribution of the Dataset

```python
print("Mean Values in the Distribution")
print("-"*35)
print(data.mean())
print("***********************************")
print("Median Values in the Distribution")
print("-"*35)
print(data.median())
```

PairPlot

```python
sns.pairplot(data, hue = 'DIVISION', vars = ['Basic Amt','Tax amt','Taxable amt','Gross Amt','Discount','Net Amt'] )
```

HeatMap

```python
plt.figure(figsize=(20,10)) 
sns.heatmap(data.corr(), annot=True)
```

Line Plot

```python
plt.style.use('seaborn')
x  = [(i+1) for i in range(14)]
y1 = data['Basic Amt'][1:15]
y2 = data['Tax amt'][1:15]
y3 = data['Gross Amt'][1:15]
y4 = data['Net Amt'][1:15]
plt.plot(x, y1, label="Basic Amt", color = 'B')
plt.plot(x, y2, label="Tax amt", color = 'R')
plt.plot(x, y2, label="Gross Amt", color = 'G')
plt.plot(x, y2, label="Net Amt", color = 'Y')
plt.plot()
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line Graph Example")
plt.legend()
plt.show()
```
